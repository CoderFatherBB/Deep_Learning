```python
import torch
import cv2

# Define scale_coords and clip_coords functions
def scale_coords(img_shape, coords, actual_shape):
    gain = min(img_shape[0] / actual_shape[0], img_shape[1] / actual_shape[1])
    pad = (img_shape[1] - actual_shape[1] * gain) / 2, (img_shape[0] - actual_shape[0] * gain) / 2
    coords[:, :4] -= torch.tensor([pad[0], pad[1], pad[0], pad[1]])
    coords[:, :4] /= gain
    clip_coords(coords, actual_shape)
    return coords

def clip_coords(coords, img_shape):
    coords[:, 0].clamp_(0, img_shape[1])  # x1
    coords[:, 1].clamp_(0, img_shape[0])  # y1
    coords[:, 2].clamp_(0, img_shape[1])  # x2
    coords[:, 3].clamp_(0, img_shape[0])  # y2

# Load the custom YOLOv6 model
checkpoint = torch.load('runs/train/exp25/weights/best_ckpt.pt')
model = checkpoint['model']  # Extract the model from the checkpoint dictionary

# Convert model parameters to float data type
model = model.float()

# Set the model to evaluation mode
model.eval()

# Set up OpenCV video capture
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        print('hi')
        break

    # Perform inference
    img = torch.from_numpy(frame).permute(2, 0, 1).float().div(255.0).unsqueeze(0)
    img = img.float()  # Convert input tensor to float data type
    pred = model(img)[0]

    # Apply non-maximum suppression
    pred = non_max_suppression(pred, conf_thres=0.5, iou_thres=0.5)[0]

    # Process detections
    for det in pred:
        # Rescale coordinates to original image size
        det[:, :4] = scale_coords(img.shape[2:], det[:, :4], frame.shape).round()

        # Draw bounding boxes and labels
        for *xyxy, conf, cls in det:
            label = f'{cls.item()} {conf:.2f}'
            plot_one_box(xyxy, frame, label=label)

    # Display the resulting frame
    cv2.imshow('Object Detection', frame)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close the OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
